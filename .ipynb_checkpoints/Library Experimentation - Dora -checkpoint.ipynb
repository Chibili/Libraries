### Reading Data & Configuration

```python
# Import required module
import pandas as pd
from Dora import Dora
  
# Create object
dora = Dora()
  
# Add dataset path as argument 
dora.configure(output = 'cost', data = 'train.csv')
  
# Display dataset
dora.data
```

## Cleaning


```python
train = pd.read_csv('train.csv')

dora = Dora(output = 0, data = train)
dora.data
```

# Imputing Missing Values

```python
dora.impute_missing_values()
dora.data
```

```python
# impute the missing values (using the average of each column)
dora.impute_missing_values()
dora.data
```
