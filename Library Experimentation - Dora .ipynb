### Reading Data & Configuration

```python
import numpy as np # linear algebra
import pandas as pd
```

```python
train=pd.read_csv("train.csv")
```

```python
from Dora import Dora
dora = Dora()  #Intialize dora
```

#### Create the Dora data set and define the target variable in the data set.
+ In the train data set, "price" is declared the target variable.

```python
dora.configure(output = 'price', data = train )
dora.data
```

```python
#Remove unwanted features from the data set

dora.remove_feature('ID')
dora.data
```

```python
#Perform one-hot encoding for categorical variables

dora.extract_ordinal_feature('SourceState')
dora.data
```

```python
dora.extract_ordinal_feature('destinationState')
dora.data
```

```python
dora.extract_ordinal_feature('vehicleType')
dora.data
```
